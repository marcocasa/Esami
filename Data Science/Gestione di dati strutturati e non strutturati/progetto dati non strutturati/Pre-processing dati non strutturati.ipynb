# DECESSI

```python
decessi = pd.read_csv('dataset/morti/DCIS_CMORTE1_EV_15052023111151999.csv')
```

```python
decessi.head()
```

```python
decessi = decessi[['Territorio', 'Sesso', 'Causa iniziale di morte - European Short List', 'TIME', 'Value']]
```

```python
decessi_finale = pd.DataFrame(columns = ['nome_malattia', 'numero_decessi_maschi', 'numero_decessi_femmine', 'anno', 'provincia'])
```

Modifico il dataset in modo da avere come colonne 'nome_malattia', 'numero_decessi_maschi', 'numero_decessi_femmine', 'anno', 'provincia' 

```python
for p in set(decessi['Territorio']):
    decessi_territorio = decessi[decessi['Territorio'] == p]
    for c in set(decessi_territorio['Causa iniziale di morte - European Short List']):
        decessi_territorio_malattia = decessi_territorio[decessi_territorio['Causa iniziale di morte - European Short List'] == c]
        decessi_territorio_malattia_maschi = decessi_territorio_malattia[decessi_territorio_malattia['Sesso'] == 'maschi']['Value'].values[0]
        decessi_territorio_malattia_femmine = decessi_territorio_malattia[decessi_territorio_malattia['Sesso'] == 'femmine']['Value'].values[0]
        row = [c, decessi_territorio_malattia_maschi, decessi_territorio_malattia_femmine, '2020', p]
        decessi_finale.loc[len(decessi_finale)] = row
        
```

```python
decessi.head(6)
```

Sostituisco il nome della provincia nei decessi con il corrispondente indice della tabella province

```python
for i in range(decessi_finale.shape[0]):
    provincia_nome = decessi_finale.loc[i, 'provincia']
    indice = province_puglia[province_puglia['nome'] == provincia_nome]['id_prov'].values[0]
    decessi_finale.loc[i, 'provincia'] = indice
```

## POPOLAZIONE COMUNI

```python
pop_maschi = pd.read_csv('dataset/popolazione/DCIS_POPRES1_15052023120359710.csv')
pop_femmine = pd.read_csv('dataset/popolazione/DCIS_POPRES1_15052023120546025.csv')
```

```python
pop_maschi.head()
```

```python
pop_femmine.head()
```

```python
comuni_puglia['pop_fem'] = None
comuni_puglia['pop_mas'] = None
comuni_puglia['anno'] = 2020
```

```python
comuni_puglia.head()
```

```python
comuni_puglia.head()
```

Merge dei dataset relativi alla popolazione femminile e alla popolazione maschile

```python
for i in range(comuni_puglia.shape[0]):
    comune = comuni_puglia.loc[i, 'nome']
    pop_maschi_comune = pop_femmine[pop_femmine['Territorio'] == comune]['Value'].values
    
    # Se non sono presenti i dati sulla popolazione
    if len(pop_maschi_comune) == 0:
        pop_maschi_comune = None
    else:
        pop_maschi_comune = pop_maschi_comune[0]
    pop_femmine_comune = pop_maschi[pop_maschi['Territorio'] == comune]['Value'].values
    if len(pop_femmine_comune) == 0:
        pop_femmine_comune = None
    else:
        pop_femmine_comune = pop_femmine_comune[0]
        
    comuni_puglia.loc[i, 'pop_mas'] = pop_maschi_comune
    comuni_puglia.loc[i, 'pop_fem'] = pop_femmine_comune
```

# STAZIONI, RILEVAZIONI E RILEVA

```python
stazioni = pd.DataFrame(columns= ['id_staz', 'nome', 'tipo_zona', 'tipo_staz', 'lat', 'long', 'id_comune'])
rileva = pd.DataFrame(columns= ['id_staz', 'id_indicatore'])
rilevazioni = pd.DataFrame(columns= ['id_rilevazione', 'anno', 'giorni_superamento', 'media_annuale', 'nome_indicatore', 'massimo', 'id_stazione'])
```

```python
rilevazioni_pm25 = pd.read_csv('dataset/indicatori/puglia_PM2_5_2020.csv', sep=';')
rilevazioni_pm10 = pd.read_csv('dataset/indicatori/puglia_PM10_2020.csv', sep=';')
```

Aggiunta del punto nella latitudine e longitudine (###### => ##.####)

```python
add_point_f = lambda x: str(x[:2]) + '.' + str(x[2:])
```

```python
rilevazioni_pm10["Latitudine"] = rilevazioni_pm10["Latitudine"].astype('string')
rilevazioni_pm10['Latitudine'] = rilevazioni_pm10['Latitudine'].apply(add_point_f)
rilevazioni_pm10["Longitudine"] = rilevazioni_pm10["Longitudine"].astype('string')
rilevazioni_pm10['Longitudine'] = rilevazioni_pm10['Longitudine'].apply(add_point_f)


rilevazioni_pm25["Latitudine"] = rilevazioni_pm25["Latitudine"].astype('string')
rilevazioni_pm25['Latitudine'] = rilevazioni_pm25['Latitudine'].apply(add_point_f)
rilevazioni_pm25["Longitudine"] = rilevazioni_pm25["Longitudine"].astype('string')
rilevazioni_pm25['Longitudine'] = rilevazioni_pm25['Longitudine'].apply(add_point_f)
```

```python
dict_columns = {'Nome stazione': 'Nome_stazione', 
                                'Valore medio annuo': 'Valore_medio_annuo',
                                'Tipo zona': 'Tipo_zona',
                                'Tipo stazione': 'Tipo_stazione',
                                'Giorni di superamento del valore limite giornaliero': 'giorni_superamento'}

rilevazioni_pm25.rename(columns=dict_columns,
                       inplace=True)
rilevazioni_pm10.rename(columns=dict_columns,
                       inplace=True)
```

```python
rilevazioni_pm25.head()
```

```python
rilevazioni_pm10.head()
```

```python
def crea_stazioni_rileva_rilevazioni(rilevazioni_indicatore, rilevazioni, stazioni, comuni_puglia, rileva, nome_indicatore):
    for i in range(rilevazioni_indicatore.shape[0]):
        #Stazioni
        id_stazione = stazioni.shape[0]+1
        nome = rilevazioni_indicatore.loc[i, 'Nome_stazione']
        tipo_zona = rilevazioni_indicatore.loc[i, 'Tipo_zona']
        tipo_stazione = rilevazioni_indicatore.loc[i, 'Tipo_stazione']
        latitudine = rilevazioni_indicatore.loc[i, 'Latitudine']
        longitudine = rilevazioni_indicatore.loc[i, 'Longitudine']

        comune = rilevazioni_indicatore.loc[i, 'Comune']
        id_comune = comuni_puglia[comuni_puglia['nome'] == comune]['id_comune'].values[0]

        #Se non è presente la stazione, la inserisco
        if len(stazioni[stazioni['nome'] == nome].values) == 0:
            stazione = [id_stazione, nome, tipo_zona, tipo_stazione, latitudine, longitudine, id_comune]
            stazioni.loc[len(stazioni)] = stazione
        else:
            id_stazione = stazioni[stazioni['nome'] == nome]['id_staz'].values[0]

        #Rileva
        rileva_row = [id_stazione, nome_indicatore]
        rileva.loc[len(rileva)] = rileva_row

        #Rilevazioni
        id_rilevaz = rilevazioni.shape[0]+1
        anno = 2020
        giorni_superamento = rilevazioni_indicatore.loc[i, 'giorni_superamento']
        massimo = rilevazioni_indicatore.loc[i, 'Massimo']
        media_annuale = rilevazioni_indicatore.loc[i, 'Valore_medio_annuo']
        rilevazione_row = [id_rilevaz, anno, giorni_superamento, media_annuale, nome_indicatore, massimo, id_stazione]
        rilevazioni.loc[len(rilevazioni)] = rilevazione_row
    
    return stazioni, rileva, rilevazioni
```

```python
# Per le rilevazioni del PM2.5
stazioni, rileva, rilevazioni = crea_stazioni_rileva_rilevazioni(rilevazioni_pm25, rilevazioni, stazioni, comuni_puglia, rileva, 'Particolato PM2.5')
# Per le rilevazioni del PM10
stazioni, rileva, rilevazioni = crea_stazioni_rileva_rilevazioni(rilevazioni_pm10, rilevazioni, stazioni, comuni_puglia, rileva, 'Particolato PM10')
```

```python
rileva.to_csv('dataset/indicatori/rileva.csv', index=False)
stazioni.to_csv('dataset/indicatori/stazioni.csv', index=False)
rilevazioni.to_csv('dataset/indicatori/rilevazioni.csv', index=False)
```
